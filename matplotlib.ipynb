```python
# Library used for data visualization
import matplotlib.pyplot as plt
```

```python
# This magic command allows plots to be displayed directly inside the Jupyter notebook
%matplotlib inline


# Import NumPy library for numerical computations
import numpy as np

# Create 11 evenly spaced values between 0 and 5 (including both 0 and 5)
x = np.linspace(0, 5, 11)

# Compute the square of each value in x (element-wise operation), meaning y = x^2
y = x ** 2


# Display the values stored in x
x
y
```

```python
# plt.figure(figsize=(2,2))         # Set the size of the figure window (width, height in inches)

plt.plot(x, y, 'r')               # Plot x versus y as a line graph using red color ('r')

plt.xlabel('X Axis Title Here')   # Set the label for the x-axis

plt.ylabel('Y Axis Title Here')   # Set the label for the y-axis

plt.title('String Title Here')    # Set the title of the plot

plt.show()                        # Display the plot output
```

```python
# plt.subplot(nrows, ncols, plot_number)  
# Divide the figure into a grid of subplots (rows, columns, position number)

plt.subplot(1,2,1)  
# Create a figure with 1 row and 2 columns, and activate the first subplot

plt.plot(x, y, 'r--')  
# Plot x versus y using a red dashed line ('r--')

plt.subplot(1,2,2)  
# Activate the second subplot in the 1x2 grid

plt.plot(y, x, 'g*-')  
# Plot y versus x using green color with star markers and a solid line ('g*-')
```

```python
# Create Figure (empty canvas)
fig = plt.figure()  
# Create a new empty figure object that will contain the plot

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  
# Add axes to the figure using normalized coordinates:
# [left, bottom, width, height] where values range from 0 to 1

# Plot on that set of axes
axes.plot(x, y, 'b')  
# Plot x versus y on the created axes using blue color ('b')

axes.set_xlabel('Set X Label')  
# Set the label for the x-axis (note the use of set_ for axes methods)

axes.set_ylabel('Set y Label')  
# Set the label for the y-axis

axes.set_title('Set Title')  
# Set the title for this axes
```

```python
# Creates blank canvas
fig = plt.figure()  
# Create a new empty figure that will contain multiple axes

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # main axes  
# Add the main axes to the figure using normalized coordinates [left, bottom, width, height]

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # inset axes  
# Add a smaller inset axes inside the figure at a different position


# Larger Figure Axes 1
axes1.plot(x, y, 'b')  
# Plot x versus y on the main axes using blue color

axes1.set_xlabel('X_label_axes1')  
# Set x-axis label for the main axes

axes1.set_ylabel('Y_label_axes1')  
# Set y-axis label for the main axes

axes1.set_title('Axes 1 Title')  
# Set title for the main axes


# Insert Figure Axes 2
axes2.plot(y, x, 'r')  
# Plot y versus x on the inset axes using red color

axes2.set_xlabel('X_label_axes2')  
# Set x-axis label for the inset axes

axes2.set_ylabel('Y_label_axes2')  
# Set y-axis label for the inset axes

axes2.set_title('Axes 2 Title')  
# Set title for the inset axes
```

```python
fig = plt.figure()  
# Create a new empty figure

ax = fig.add_axes([0,0,1,1])  
# Add a full-size axes to the figure using normalized coordinates [left, bottom, width, height]

ax.plot(x, x**2, label="x**2")  
# Plot x squared (x^2) versus x and assign a label for the legend

ax.plot(x, x**3, label="x**3")  
# Plot x cubed (x^3) versus x and assign a label for the legend

ax.legend()  
# Display the legend on the plot to distinguish between the two lines
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  
# Create a figure with 1 row and 3 subplots (side by side) and set the figure size

axes[0].plot(x, x**2, x, x**3)  
# Plot x^2 and x^3 on the first subplot using default axis scaling

axes[0].set_title("default axes ranges")  
# Set title for the first subplot


axes[1].plot(x, x**2, x, x**3)  
# Plot x^2 and x^3 on the second subplot

axes[1].axis('tight')  
# Automatically adjust axis limits to tightly fit the data

axes[1].set_title("tight axes")  
# Set title for the second subplot


axes[2].plot(x, x**2, x, x**3)  
# Plot x^2 and x^3 on the third subplot

axes[2].set_ylim([0, 60])  
# Manually set the y-axis limits from 0 to 60

axes[2].set_xlim([2, 5])  
# Manually set the x-axis limits from 2 to 5

axes[2].set_title("custom axes range")  
# Set title for the third subplot
```
